```python
import seaborn
import json
import numpy as np
```

```python
with open("AddaAttN_bw.json", 'r') as f:
    AddAttN = json.load(f)
with open("SANET_bw.json", 'r') as f:
    SANET = json.load(f)
with open("AdaIN_bw.json", 'r') as f:
    AdaIN = json.load(f)
```

```python
def get_heatmap(json_file, field):
    ssim = {}
    for k in json_file:
        s, c = k[:-4].split("_")
        s = int(s.removeprefix("style"))
        c = int(c.removeprefix("content"))
        if field == "adapted rand error":
            ssim[(s,c)] =  json_file[k]['adapted rand error']["are"]
            
        elif field == "variation of information":
            ssim[(s,c)] =  json_file[k]['variation of information']["H(cs|c)"]
        
        else: 
            ssim[(s,c)] = json_file[k][field]

    ks = ssim.keys()
    ks = sorted(ks, key=lambda x : x[1])
    ks = sorted(ks, key=lambda x : x[0])
    
    ssim_list = []
    current = []
    s_idx = 0
    for k in ks:
        if s_idx != k[0]:
            ssim_list.append(current)
            current = []
            s_idx = k[0]
        current.append(ssim[k])
    ssim_list.append(current)
    ssim_arr = np.array(ssim_list)
    seaborn.heatmap(ssim_arr)
    print("Mean:", np.mean(ssim_arr))
    
```

```python
get_heatmap(AddAttN, "variation of information")
```

```python
get_heatmap(SANET, "variation of information")
```

```python
get_heatmap(AdaIN, "variation of information")
```

```python
%matplotlib inline
from PIL import Image
import numpy as np
from os import listdir
from sys import argv
import seaborn as sns
import matplotlib.pyplot as plt

dirname = "AddaAttN_out"
out_ext = "png"
chi2_distance = lambda a, b: 0.5 * np.sum([((a - b) ** 2) / ((a + b) + 1e-20)])
fig, ax = plt.subplots(nrows=3, figsize=(12, 7), sharex=True)

s = listdir("style")[0]
c = listdir("content")[0]
cs_name = f"{s[:-4]}_{c[:-4]}.{out_ext}"
print(cs_name)
cs_image = np.array(Image.open(f"{dirname}/{cs_name}").convert('RGB'))
style_image = np.array(Image.open(f"style/{s}").convert('RGB'))
chi2_distance = lambda a, b: 0.5 * np.sum([((a - b) ** 2) / ((a + b) + 1e-20)])
dist = []
sp = 1
rgb = ["red", "green", "blue"]
for i in range(3):

    h, bin_edges = np.histogram(cs_image[:, :, i], bins=np.arange(0,256), density=True)
    ax[i].bar(bin_edges[:-1],h, color="#F8B195")
    h2, bin_edges2 = np.histogram(style_image[:, :, i], bins=np.arange(0, 256), density=True)
    ax[i].bar(bin_edges2[:-1],h2, alpha=0.3, color=rgb[i])
    ax[i].set_ylabel(rgb[i])
    dist.append(chi2_distance(h, h2))
    plt.tight_layout()
    if i == 2:
        ax[i].set_xlabel("pixel value")
print(np.mean(dist))
```
